```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
documents=SimpleDirectoryReader("data").load_data()
```

```python
documents
```

```python
system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided. You also need to provide verifiable references to source content along with page number as a context
Also remove stop words or additional special characters like backslash n and make it clean
"""
## Default format supportable by LLama2
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python

!huggingface-cli login
```

```python

import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="tiiuae/falcon-7b-instruct",  # Use Falcon 7B for Q&A
    model_name="tiiuae/falcon-7b-instruct",      # Free, open-source model
    device_map="auto",
    model_kwargs={"torch_dtype": torch.float16, "load_in_8bit": True},
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.legacy.embeddings.huggingface import HuggingFaceEmbedding

# === Step 3: Initialize HuggingFace Embedding Model ===
embed_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
from llama_index.core import Settings
Settings.llm = llm
Settings.embed_model = embed_model
Settings.chunk_size = 1024
```

```python
index=VectorStoreIndex.from_documents(documents)
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("What are the key components of emotion?")
response
```

```python
import re
def clean_text(text):
    # Remove newlines, extra spaces, and special characters
    text = re.sub(r"\\n|\\r", " ", text)  # Remove \n and \r
    text = re.sub(r"[^a-zA-Z0-9 .,]", "", text)  # Remove special characters except for ., and space
    text = re.sub(r"\s+", " ", text)  # Replace multiple spaces with one space
    return text.strip()
def process_queries(queries_file, output_file):
    # Load queries from JSON
    with open(queries_file, 'r') as f:
        queries = json.load(f)

    results = []
    for query in tqdm(queries, desc="Processing queries"):
        query_id = query["query_id"]
        question = query["question"]
        print(f"Processing Query ID {query_id}: {question}")

        # Retrieve response and relevant nodes
        response = query_engine.query(question)
        context_nodes = response.source_nodes

        # Extract cleaned context and references
        context_list = []
        references = []
        for node in context_nodes:
            text = clean_text(node.node.text)
            page = node.node.metadata.get("page_label", node.node.metadata.get("page", "N/A"))
            context_list.append(f"(Page {page}) {text}")
            references.append(page)

        combined_context = "\n".join(context_list)
        cleaned_answer = clean_text(str(response))

        # Append results
        results.append({
            "ID": query_id,
            "context": combined_context,
            "answer": cleaned_answer,
            "references": json.dumps({"pages": list(set(references))})
        })
        print(results)
    # Write results to submission.csv
    df = pd.DataFrame(results)
    df.to_csv(output_file, index=False)
    print(f"Submission file saved to {output_file}")

# === Step 9: Generate Submission File ===
queries_file = "queries.json"  # Path to queries.json
output_file = "submission2.csv"  # Output submission file
process_queries(queries_file, output_file)
```
